```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\juned\Videos\data analytics\end to end projects\Healthcare Dataset\Data\Cleaned_data.csv")
```

```python
sns.barplot(data =df, x = 'Medical Condition',y = 'Billing Amount', estimator = np.mean)
plt.ticklabel_format(style='plain', axis='y')
plt.title('Average Billing Amount by Medical Condition')
plt.show()
```

```python
pd.set_option('display.max_columns', None)
```

```python
top10hospital = (df.groupby('Hospital')['Billing Amount'].sum().sort_values(ascending = False).head(10).copy().reset_index())
sns.barplot(data = top10hospital, x = 'Billing Amount', y = 'Hospital')
plt.title('Top 10 Hospitals by Revenue')
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

```python
top10doctor = (df.groupby('Doctor')['Name'].nunique().sort_values(ascending = False).head(10).copy().reset_index())
sns.barplot(data = top10doctor, x = 'Name', y = 'Doctor')
plt.title('Top 10 Doctors by Patient Count')
plt.xlabel('Patient count')
plt.show()
```

```python
top10insurance = (df.groupby('Insurance Provider')['Name'].nunique().sort_values(ascending = False).head(10).copy().reset_index())
sns.barplot(data = top10insurance, x = 'Name', y = 'Insurance Provider')
plt.title('Insurance Provider Distribution')
plt.xlabel('Patient count')
plt.show()
```

```python
plt.pie(df['Gender'].value_counts(), labels = df['Gender'].value_counts().index, 
    explode=[0.05, 0.05], autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
print(df['Gender'].value_counts())
```

```python
blood = (df.groupby('Blood Type')['Name'].nunique().sort_values(ascending = True).copy().reset_index())
sns.barplot(data = blood, x = 'Blood Type' , y = 'Name')
plt.title('Blood Type Distribution')
plt.ylabel('Patient count')
plt.show()
print(blood)
```

```python
monthly = (df.groupby('Month of Admission')['Name'].nunique().reset_index())
sns.lineplot(data=monthly, x = 'Month of Admission', y = 'Name', marker = 'o')
plt.title('Monthly Admissions Trend')
plt.ylabel('Patient count')
plt.show()
```

```python
monthlybilling = (df.groupby('Month of Admission')['Billing Amount'].sum().sort_values(ascending = True).copy().reset_index())
sns.lineplot(data=monthlybilling, x = 'Month of Admission', y = 'Billing Amount', marker = 'o')
plt.ticklabel_format(style='plain', axis='y')
plt.title('Monthly Billing Trend')
plt.xlabel('Month')
plt.show()
```

```python
average = (df.groupby('Medical Condition')['Length_of_Stay'].mean().reset_index())
sns.barplot(data = average, x = 'Medical Condition', y = 'Length_of_Stay')
plt.title('Average Length of Stay by Medical Condition')
plt.ylabel('Length Of Stay')
plt.show()
```

```python
sns.countplot(data=df, x = 'Age_Group')
plt.title('Age Group Distribution')
plt.ylabel('Patient count')
plt.show()
```

```python
sns.countplot(data = df, x = 'Admission Type')
plt.title('Admission Type Distribution')
plt.ylabel('Patient count')
plt.show()
```

```python
sns.boxplot(data = df, x = 'Age_Group' , y = 'Billing Amount')
plt.title('Billing Amount by Age Group')
plt.show()
```

```python
plt.pie(df['Test Results'].value_counts(), labels = df['Test Results'].value_counts().index, 
    explode=[0.05, 0.05, 0.05], autopct='%1.1f%%')
plt.title('Test Results Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=df,x='Length_of_Stay',y='Billing Amount')
plt.title("Billing Amount Distribution by Length of Stay")
plt.xlabel("Length of Stay (Days)")
plt.ylabel("Billing Amount")
plt.show()
```
