```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
```

```python
df['Day of Admission'] = df['Date of Admission'].dt.day
df['Month of Admission'] = df['Date of Admission'].dt.month
df['Year of Admission'] = df['Date of Admission'].dt.year
```

```python
df['Day of Discharge'] = df['Discharge Date'].dt.day
df['Month of Discharge'] = df['Discharge Date'].dt.month
df['Year of Discharge'] = df['Discharge Date'].dt.year
```
